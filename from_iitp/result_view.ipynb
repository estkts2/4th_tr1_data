```python
import moviepy.editor  as mpe
import imageio
from glob import glob
from tqdm.auto import tqdm
import IPython.display as ipd
import cv2
start = 0 
#end  = start + 5000
end = -1
 
imgs = sorted(glob('last_infer/*.jpg'))[start:end]
imgs = [imageio.imread(e) for e in tqdm(imgs)]
imgs = [cv2.resize(e, (960, 540), cv2.INTER_AREA) for e in imgs] 
vc =mpe.ImageSequenceClip(imgs, fps=3)
vc.write_videofile('out.mp4')
ipd.Video('out.mp4')
```
